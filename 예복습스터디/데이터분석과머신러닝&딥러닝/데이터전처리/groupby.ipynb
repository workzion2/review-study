```python
import pandas as pd

# 데이터 생성
df = pd.DataFrame({
    'Category': ['Fruit', 'Fruit', 'Veggie', 'Fruit', 'Veggie'],
    'Name': ['Apple', 'Banana', 'Carrot', 'Apple', 'Broccoli'],
    'Sales': [10, 20, 15, 30, 5]
})

# 'Category'별로 묶어서 'Sales'의 합계 구하기
result = df.groupby('Category')['Sales'].sum()

print(result)
# 출력 결과:
# Category
# Fruit     60
# Veggie    20
```
